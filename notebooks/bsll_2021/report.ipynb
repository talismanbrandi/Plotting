```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import uproot
```

```python
file = uproot.open('/home/apaul/beegfs/TheNewHope/VFNCULO-1/SM/PMD/SM/MCout.root')
tree = file['StandardModel_Observables']
df = pd.DataFrame()
df['Regtilde_1_LQ4'] = tree.get('Regtilde_1_LQ4').array(library="np")
df['Regtilde_2_LQ4'] = tree.get('Regtilde_2_LQ4').array(library="np")
df['P_5p_LQ1_LHCb'] = tree.get('P_5p_LQ1_LHCb').array(library="np")

df_PMD = df.copy(deep=True)


file = uproot.open('/home/apaul/beegfs/TheNewHope/VFNCULO-1/SM/PDD/SM/MCout.root')
tree = file['StandardModel_Observables']
df = pd.DataFrame()
df['Regtilde_1_LQ4'] = tree.get('Regtilde_1_LQ4').array(library="np")
df['Regtilde_2_LQ4'] = tree.get('Regtilde_2_LQ4').array(library="np")
df['P_5p_LQ1_LHCb'] = tree.get('P_5p_LQ1_LHCb').array(library="np")

df_PDD = df.copy(deep=True)

file = uproot.open('/home/apaul/beegfs/TheNewHope/VFNCULO-1/SM/FDD/SM/MCout.root')
tree = file['StandardModel_Observables']
df = pd.DataFrame()
df['Regtilde_1_LQ4'] = tree.get('Regtilde_1_LQ4').array(library="np")
df['Regtilde_2_LQ4'] = tree.get('Regtilde_2_LQ4').array(library="np")
df['P_5p_LQ1_LHCb'] = tree.get('P_5p_LQ1_LHCb').array(library="np")

df_FDD = df.copy(deep=True)
```

```python
df_PMD['sum_1_2'] = df_PMD['Regtilde_1_LQ4'] + df_PMD['Regtilde_2_LQ4']
df_PMD['diff_1_2'] = df_PMD['Regtilde_1_LQ4'] - df_PMD['Regtilde_2_LQ4']

df_PDD['sum_1_2'] = df_PDD['Regtilde_1_LQ4'] + df_PDD['Regtilde_2_LQ4']
df_PDD['diff_1_2'] = df_PDD['Regtilde_1_LQ4'] - df_PDD['Regtilde_2_LQ4']

df_FDD['sum_1_2'] = df_FDD['Regtilde_1_LQ4'] + df_FDD['Regtilde_2_LQ4']
df_FDD['diff_1_2'] = df_FDD['Regtilde_1_LQ4'] - df_FDD['Regtilde_2_LQ4']
```
